```python
#Installing and setting up the required files and libraries for SQL environment
!pip install pandasql
import pandasql as ps
import sqlite3 as sql
```

```python
#5 How to use Pysql in Google Colab and also how to give queries
#Generating a Query
query_1 = '''
SELECT
  facility_zip as ZIP,
  AVG(SCORE) AS avg_score,
  MIN(SCORE) AS min_score,
  MAX(SCORE) AS max_score
FROM
  health_scores
GROUP BY
  facility_zip
ORDER BY
  avg_score DESC
'''
```

```python
#Summarising the data
summary_health_scores = ps.sqldf(query_1,globals()) #Running the query
summary_health_scores['ZIP'] = summary_health_scores['ZIP'].astype(int) #Converting into int
summary_health_scores#Viewing the data
```

```python
#Generating a Query
#6 Generate to me a sql query to summarise a column ZIP
query_2 = '''
SELECT
  "Zip Code" AS ZIP,
  AVG("SITE UNITS") AS avg_site_units,
  MAX("SITE UNITS") AS max_site_units,
  MIN("SITE UNITS") AS min_site_units,
  COUNT(*) AS num_sites,
  AVG("PROJECT TOTAL UNITS") AS avg_project_units,
  MAX("PROJECT TOTAL UNITS") AS max_project_units,
  MIN("PROJECT TOTAL UNITS") AS min_project_units,
  COUNT("PROJECT TOTAL UNITS") AS num_project_units,
  AVG("LAHD FUNDED") AS avg_lahd,
  MAX("LAHD FUNDED") AS max_lahd,
  MIN("LAHD FUNDED") AS min_lahd,
  SUM("LAHD FUNDED") AS sum_lahd,
  COUNT("LAHD FUNDED") AS num_lahd,
  AVG("LEVERAGE") AS avg_leverage,
  MAX("LEVERAGE") AS max_leverage,
  MIN("LEVERAGE") AS min_leverage,
  SUM("LEVERAGE") AS sum_leverage,
  COUNT("LEVERAGE") AS num_leverage,
  AVG("TDC") AS avg_tdc,
  MAX("TDC") AS max_tdc,
  MIN("TDC") AS min_tdc,
  SUM("TDC") AS sum_tdc,
  COUNT("TDC") AS num_tdc
FROM housing_prices
GROUP BY "Zip Code"
ORDER BY avg_site_units DESC
'''
```

```python
#Running the query and then viewing the same
summary_housing_data = ps.sqldf(query_2,globals())
summary_housing_data
```

```python
#In this query we are joining the data based on the common zip codes present in both the summarized data files
#and then we can use the data set t operform a detailed analysis
#7 How to join in SQL by using inner join
query_to_join = '''
SELECT
  summary_health_scores.ZIP,
  summary_health_scores.avg_score,
  summary_housing_data.avg_site_units,
  summary_housing_data.avg_project_units,
  summary_housing_data.avg_lahd,
  summary_housing_data.avg_leverage,
  summary_housing_data.avg_tdc

FROM
  summary_health_scores
INNER JOIN summary_housing_data
ON summary_health_scores.ZIP = summary_housing_data.ZIP
'''
```

```python
#Here we are merging the files together using SQL  and saving it into a new data frame named merged
merged = ps.sqldf(query_to_join,locals())
merged #Viewing the merged file
```

```python
#Performing analysis on the final data set
#Plotting the values to get a better understanding of the distribution
plt.scatter(merged['ZIP'],merged['avg_score'],color = 'b')
plt.xlabel('ZIP Codes')
plt.ylabel('Average Health Scores')
plt.title('Average Health Scores by ZIP Code')
plt.show()
```
