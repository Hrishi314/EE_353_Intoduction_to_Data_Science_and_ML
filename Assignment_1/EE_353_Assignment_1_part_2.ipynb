```python
#Assignment Part 2
#Importing all the necessary Library
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#The problem statement and Hypothesis
'''
The data given gives us the number of users who are having Wireless and Wireline connections in India from 2014 to 2022
in  a batch of 3 months
Hypothesis :-
The number of Wireless users will be much more than the number of Wireline users, and the difference between
the number of Wireless usersand Wireline users will increase with time as Wireless is far more superior trechnology
as compared to Wireline technology and is convienient as well.
'''
```

```python
#Reading and Storing the data
data_frame = pd.read_csv("/content/drive/MyDrive/EE_353_Data/Assignment_1/Month-wise_BroadbandSubscribers-Wireline_vs_Wireless-April_2014-to-Dec_2022.csv")
```

```python
#Viewing the data
data_frame
```

```python
#Plotting the data in order to get a better understanding of the data given
plt.scatter(data_frame['S. No.'],data_frame['Wireless Subscribers - Cumulative (Crore)'],color = 'blue')
plt.scatter(data_frame['S. No.'],data_frame['Wireline Subscribers - Cumulative (Crore)'],color ='red')
```

```python
'''
In order to fix the missing values we can run the following logic
We just need to take the consecutive differences between the elements of Wireless Subscribers and then compare it with
a standard given value in order to swap the values.I chose the value as 0 as it is quite obvious that
the number of Wireless users is far more than the number of Wireline users.

But since this is a small dataset,we can actually do the swapping manually and then
'''
```

```python
#Implementing the above logic
for i in range(1,len(data_frame)):
  if data_frame.loc[i,'Wireless Subscribers - Cumulative (Crore)'] - data_frame.loc[i,'Wireline Subscribers - Cumulative (Crore)'] < 0:
    data_frame.loc[i,'Wireless Subscribers - Cumulative (Crore)'],data_frame.loc[i,'Wireline Subscribers - Cumulative (Crore)'] = data_frame.loc[i,'Wireline Subscribers - Cumulative (Crore)'],data_frame.loc[i,'Wireless Subscribers - Cumulative (Crore)']
```

```python
#Plotting the data again in order to check if the above logic worked
plt.scatter(data_frame['S. No.'],data_frame['Wireless Subscribers - Cumulative (Crore)'],color = 'blue')
plt.scatter(data_frame['S. No.'],data_frame['Wireline Subscribers - Cumulative (Crore)'],color ='red')
```

```python
'''
Yup, we have successfully done the preprocessing part !!!
Now the data is good to be analyzed
'''
```

```python
#Analyzing the data

#Plotting the given data values
plt.plot(data_frame['S. No.'],data_frame['Wireless Subscribers - Cumulative (Crore)'],color = 'blue')
plt.plot(data_frame['S. No.'],data_frame['Wireline Subscribers - Cumulative (Crore)'],color ='red')
#Printing the plot
plt.show()
```

```python
#Making a new column for years which is more of a continuous type by using the given data set
for i in range(len(data_frame)):
  data_frame['Year_continuous'] = 2014 + 0.25*data_frame['S. No.']
```

```python
#Calculating the gradients of the given numbers
data_frame['Slope_Wireless'] = 0
data_frame['Slope_Wireline'] = 0
for j in range(1,len(data_frame)):
  data_frame.at[j,'Slope_Wireless'] = (data_frame['Wireless Subscribers - Cumulative (Crore)'][j] - data_frame['Wireless Subscribers - Cumulative (Crore)'][j-1])/(data_frame['Year_continuous'][j] - data_frame['Year_continuous'][j-1])
  data_frame.at[j,'Slope_Wireline'] = (data_frame['Wireline Subscribers - Cumulative (Crore)'][j] - data_frame['Wireline Subscribers - Cumulative (Crore)'][j-1])/(data_frame['Year_continuous'][j] - data_frame['Year_continuous'][j-1])
  data_frame.at[j,'Slope_total'] = (data_frame['Total Subscribers - Cumulative (Crore)'][j] - data_frame['Total Subscribers - Cumulative (Crore)'][j-1])/(data_frame['Year_continuous'][j] - data_frame['Year_continuous'][j-1])
```

```python
#Viewing the data frame
data_frame
```

```python
#Plotting the slopes in order to get a better idea of the rate of change of variables
plt.plot(data_frame['Year_continuous'],data_frame['Slope_Wireless'],color = 'blue')
plt.plot(data_frame['Year_continuous'],data_frame['Slope_Wireline'],color ='red')
plt.plot(data_frame['Year_continuous'],data_frame['Slope_total'],color ='green')
plt.show()
```

```python
#Calculating the maximum slopes of a given column
max_slope_wireless = data_frame['Slope_Wireless'].max()
max_slope_wireline = data_frame['Slope_Wireline'].max()
max_slope_total = data_frame['Slope_total'].max()

#Doing the calculations required and printing the values

for j in range(len(data_frame)):
  if data_frame['Slope_Wireless'][j] == max_slope_wireless:
    print(data_frame['Year_continuous'][j])
    print(max_slope_wireless)

  if data_frame['Slope_Wireline'][j] == max_slope_wireline:
    print(data_frame['Year_continuous'][j])
    print(max_slope_wireline)

  if data_frame['Slope_total'][j] == max_slope_total:
    print(data_frame['Year_continuous'][j])
    print(max_slope_total)
```
